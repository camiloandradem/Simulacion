## Lucas Numbers

```python
import numpy as np

#a esta funcion le entra la cantidad de numeros que se quieren calcular

def lucasNumbers(n):
    
    lucas = []

    lucas.append(1)
    lucas.append(3)
    
    for i in range (1,n-1):
        
        lucas.append(lucas[i-1]+lucas[i])
        
    return lucas

```

```python
#pruebo calculando 15 numeros

lucasNumbers(15)
```

```python
import numpy as np
import time

#Este calcula numeros de lucas durante la cantidad de segundos que se ingresen en la funcion

def lucasNumbersTime(n):
    lucas = []

    lucas.append(1)
    lucas.append(3)

    hr = time.mktime(time.localtime())
    l = hr + n

    i=0

    while (hr<l):
        hr = time.mktime(time.localtime())
        lucas.append(lucas[i-1]+lucas[i])
        i += 1
    
    return lucas
```
